# libraries for the implementation of below functions


```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
import skimage.io
```

# Function for displaying the image


```python
def load_image(image):                           
    plt.imshow(image)   
    plt.show()
```

# Function for saving the image

```python
def save_image(image,image_name):                      # the user is required to enter the desired file name as image_name as which he/she wants to save the image              
    skimage.io.imsave(fname=image_name, arr=image)
    print('Image Saved as {f}'.format(f=image_name))
    load_image(image)
```

# Function for inverting the image

```python
def invert_image(image):                         
    t=np.invert(image)
    return t
```

# Function for converting the image to black and white

```python
def black_and_white(image):
    grarray=greyscale(image)
    bw = (grarray > grarray.mean())*255
    return bw
```

# Function for converting the image to grayscale


```python
# This function uses Luminosity formula for calculation of R,G,B values
def greyscale(img):
    grayImage = np.zeros(img.shape)
    R = np.array(img[:, :, 0])
    G = np.array(img[:, :, 1])
    B = np.array(img[:, :, 2])

    R = (R *0.2126)
    G = (G *0.7152)
    B = (B *0.0722)

    Avg = (R+G+B)
    grayImage = img.copy()

    for i in range(3):
        grayImage[:,:,i] = Avg
    return grayImage
```
