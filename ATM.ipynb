# 1. ATM Class

```python
class ATM:  
    
    c=10001                                                       # static variable c for initializng account number for client
    d={10001:['cara',7894]}                                       # static dictionary for keeping name and pin of user with the respective account no.
    
    def __init__(self,accounts_no=None):
        self.__balance=0
        self.__checking_balance=551
        if accounts_no is None:
            ATM.c+=1
            self.__account_no=ATM.c
        else:
            self.__account_no=accounts_no
    
    # Getters and Setters for private variables
    
    def get_ac_no(self):                                          # to get account number of client
        return self.__account_no
    def get_balance(self):                                        # to get balance of the client
        return self.__balance
    def set_balance(self,amount,transaction_type="Withdrawal"):   # to set balance when the amount is deposited or withdrawed     
        if transaction_type=='DEPOSIT':
            self.__balance+=amount
        else:
            self.__balance-=amount
    def get_checking_balance(self):                               # to get amount from checking balance account
        return self.__checking_balance
    def set_checking_balance(self,amount):                        # to set amount in checking balance amount when the amount is deducted from it
        if amount<=self.__checking_balance:
            self.__checking_balance-=amount
            return 0
        else:
            print("Insufficient balance in checking account!!")
            return 1

  # class Methods

    def register(self,name,pin):                                  # Function for registration of new user
        ATM.d[self.get_ac_no()]=[name,pin]
        
    def check_pin(self,input_pin):                                # Function for authenticating the pin
        flag=0
        if ATM.d[self.get_ac_no()][1]==input_pin:
            print("WELCOME {s}".format(s=ATM.d[self.get_ac_no()][0]))
        else:
            flag+=1
            print("Incorect pin!")
        return flag 
    
    def ATM_menu_for_user(self):                                  # function for selection of operations by the user
        f=True
        while(f==True):
            print("Please select (1-5) for the following options: ")
            print("1. Account Balance")
            print("2. Cash Deposit")
            print("3. Cash Withdrwal")
            print("4. Transfer Funds")
            print("5. Exit")
            t=int(input())

            if t==1:
                print("Account Balance: ",self.get_balance())
            elif t==2:
                print("Enter the amount:")
                amount=int(input())
                self.set_balance(amount,'DEPOSIT')
                print("Account Balance: ",self.get_balance())
            elif t==3:
                print("Enter the amount:")
                amount=int(input())
                if amount<=self.get_balance():
                    self.set_balance(amount)
                    print("Account Balance: ",self.get_balance())
                else:
                    print("Insufficient Balance!")
            elif t==4:
                print("enter the amount to be transfered")
                m=int(input())
                if self.set_checking_balance(m)==0:
                    self.set_balance(m,'DEPOSIT')
                    print("Account Balance: ",self.get_balance())
                    print("Checking Account Balance: ",self.get_checking_balance())
            elif t==5:
                f=False
                print("THANK YOU!")
                break
            else:
                f=False
                print("INVALID CHOICE!!!")
                break
                    
```

# 2. Functions for account login and account creation


```python
def account_login():                  # the user will be required to enter his/her account number and pin for authentication
    print('Enter account number: ')
    ac_no=int(input())
    print('Enter Pin: ')
    pin=int(input())
    client=ATM(ac_no)
    if pin==client.d[client.get_ac_no()][1]:
        print("Please enter the pin again")
        p=int(input())
        if client.check_pin(p)==0:
            client.ATM_menu_for_user()
        else:
            print("Pins don't match!!")
    else:
        print('Invalid credentials!!')
        
def account_register():            # the user will be required to give his/her name and the pin that he/she wants to set and an account will be created 
    print('Enter Name: ')
    name=input()
    print('Enter your 4 digit Pin: ') 
    pin=int(input())
    client=ATM()
    client.register(name,pin)
    print("Congratulations, ACCOUNT CREATED!")
    print("Account Number: ",client.get_ac_no())
    
```
